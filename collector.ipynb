```python
import os
import pickle
import pandas as pd

# Directory containing the pickle files
jobs_dir = 'jobs'

# List to hold the data
data_list = []

# Iterate over all pickle files in the directory
for filename in os.listdir(jobs_dir):
    if filename.endswith('.pickle'):
        with open(os.path.join(jobs_dir, filename), 'rb') as handle:
            data = pickle.load(handle)
            data_dict = {
                'job_id': data.job_id,
                'title': data.title,
                'company': data.company,
                'place': data.place,
                'date': data.date,
                'description': data.description,
                'seniority_level': getattr(data, 'seniority_level', None),
                'employment_type': getattr(data, 'employment_type', None),
                'industries': getattr(data, 'industries', None),
                'job_function': getattr(data, 'job_function', None),
                'link': data.link
            }
            data_list.append(data_dict)  # Manually create a dictionary from EventData attributes

# Create a DataFrame from the list of dictionaries
df = pd.DataFrame(data_list)

# Save the DataFrame to a CSV file
df.to_csv('jobs_data.csv', index=False)

print("Data has been written to jobs_data.csv")
```

```python
import os
import logging
import json
from linkedin_jobs_scraper import LinkedinScraper
from linkedin_jobs_scraper.events import Events, EventData
from linkedin_jobs_scraper.query import Query, QueryOptions

# Set up logging
logging.basicConfig(level=logging.INFO)

# Create 'jobs' directory if it doesn't exist
jobs_dir = 'jobs'
os.makedirs(jobs_dir, exist_ok=True)

# Callback when job data is retrieved
def on_data(data: EventData):
    logging.info(f"Writing job ID: {data.job_id}")
    job_details = {
        'job_id': data.job_id,
        'title': data.title,
        'company': data.company,
        'location': data.place,
        'description': data.description,
        'link': data.link
    }
    with open(f'{jobs_dir}/{data.job_id}.json', 'w') as json_file:
        json.dump(job_details, json_file, indent=4)

# Callback for errors
def on_error(error):
    logging.error(f"Error occurred: {error}")

# Callback when scraping ends
def on_end():
    logging.info("Scraping ended.")

# Initialize the scraper
scraper = LinkedinScraper(
    chrome_executable_path=None,  # Ensure ChromeDriver is in PATH if None
    chrome_binary_location=None,
    chrome_options=None,
    headless=True,
    max_workers=1,
    slow_mo=15,
    page_load_timeout=30
)

# Attach event listeners
scraper.on(Events.DATA, on_data)
scraper.on(Events.ERROR, on_error)
scraper.on(Events.END, on_end)

# Define search queries for multiple roles
queries = [
    Query(query='Business Analyst', options=QueryOptions(locations=['India'], limit=1000)),
    Query(query='Data Analyst', options=QueryOptions(locations=['India'], limit=1000)),
    Query(query='Data Scientist', options=QueryOptions(locations=['India'], limit=1000)),
]

# Run the scraper
try:
    scraper.run(queries)
except Exception as e:
    logging.error(f"Failed to complete scraping: {e}")
```

```python
# Flask App to Display Job Listings

import os
import json
from flask import Flask, render_template_string

app = Flask(__name__)

# HTML template for job listings
HTML_TEMPLATE = '''
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Job Listings</title>
    <style>
        body { font-family: Arial, sans-serif; background-color: #f8f9fa; margin: 0; padding: 20px; }
        h1 { text-align: center; color: #333; }
        .job-card { background-color: white; margin: 10px auto; padding: 15px; box-shadow: 0 0 10px rgba(0, 0, 0, 0.1); width: 80%; border-radius: 8px; }
        .job-title { font-size: 1.5em; color: #007bff; }
        .job-company { font-weight: bold; color: #555; }
        .job-location { color: #888; margin: 5px 0; }
        .job-description { margin: 10px 0; }
        .job-link { display: inline-block; padding: 5px 10px; background-color: #007bff; color: white; text-decoration: none; border-radius: 5px; }
        .job-link:hover { background-color: #0056b3; }
    </style>
</head>
<body>
    <h1>Latest Job Listings</h1>
    {% if jobs %}
        {% for job in jobs %}
        <div class="job-card">
            <div class="job-title">{{ job['title'] }}</div>
            <div class="job-company">{{ job['company'] }}</div>
            <div class="job-location">{{ job['location'] }}</div>
            <div class="job-description">{{ job['description'][:300] }}...</div>
            <a class="job-link" href="{{ job['link'] }}" target="_blank">View on LinkedIn</a>
        </div>
        {% endfor %}
    {% else %}
        <p style="text-align: center; color: #888;">No job listings available.</p>
    {% endif %}
</body>
</html>
'''

# Load job data from JSON files in 'jobs' folder
def load_jobs():
    jobs = []
    jobs_folder = 'jobs'
    if os.path.exists(jobs_folder):
        for file in os.listdir(jobs_folder):
            if file.endswith('.json'):
                with open(os.path.join(jobs_folder, file), 'r') as json_file:
                    try:
                        job_data = json.load(json_file)
                        jobs.append(job_data)
                    except json.JSONDecodeError:
                        continue
    return jobs

# Route to display job listings
@app.route('/')
def job_listings():
    jobs = load_jobs()
    return render_template_string(HTML_TEMPLATE, jobs=jobs)

# Run the Flask app
if __name__ == '__main__':
    app.run(debug=True, host='0.0.0.0', port=5000)
```
